```python
import numpy as np
from math import factorial
import matplotlib.pyplot as plt
```

### Implementing Taylor Approximations with Code

For evaluating values and plotting graphs it can be helpful to make a function that returns the Taylor polynomial of a specified degree.  Here is a function that gives the Taylor polynomial, centered at $x_0 = 0$ for $f(x) = e^x.$

```python
# Define Taylor Series approximations around x0 = 0
def taylor_exp_approximation_near_0(x, degree):
    x0 = 0  # Center point of the Taylor Series
    num_terms = degree + 1  # Number of terms to use in the approximation
    approximation = 0
    for n in range(num_terms):
        coefficient = 1 / factorial(n)
        term = coefficient * (x - x0)**n
        approximation += term
    return approximation
```

Here's an example showing how to estimate the value of $f(1) = e^1 = e$ using the tenth degree Taylor polynomial:

```python
taylor_exp_approximation_near_0(1,10)
```

We can also pass in a Numpy array of points and the function will return an array of approximations:

```python
x = np.linspace(-2,2,100)
y = taylor_exp_approximation_near_0(x,10)

print(x[:5])
print(y[:5])
```

We can make use of the arrays to make a plot:

```python
plt.plot(x,y)
```

Here's some example code for plotting the $e^x$ and multiple Taylor polynomial approximations.  You can mimic this code to help with the exercises.

```python


# Define the exponential function
def exp_function(x):
    return np.exp(x)

# Define Taylor Series approximations around x0 = 0
def taylor_exp_approximation_near_0(x, degree):
    x0 = 0  # Center point of the Taylor Series
    num_terms = degree + 1  # Number of terms to use in the approximation
    approximation = 0
    for n in range(num_terms):
        coefficient = 1 / factorial(n)
        term = coefficient * (x - x0)**n
        approximation += term
    return approximation

# Generate x values
x = np.linspace(-2, 2, 500)

# Compute the true exponential function values
true_exp = exp_function(x)

# Plot the true exponential function
plt.figure(figsize=(8, 5))
plt.plot(x, true_exp, label="True $e^x$", color="black", linewidth=2)

# Plot Taylor Series approximations for various orders
degrees = [0, 1, 2, 3, 4]  # Orders of approximation
colors = ["red", "blue", "green", "orange", "purple"]

for degree, color in zip(degrees, colors):
    approximation = taylor_exp_approximation_near_0(x, degree)
    plt.plot(x, approximation, label=f"Degree {degree} Approximation", linestyle="--", color=color)

# Customize the plot
plt.title("Taylor Series Approximations of $e^x$ Near $x = 0$", fontsize=16)
plt.xlabel("$x$", fontsize=14)
plt.ylabel("$y$", fontsize=14)
plt.legend(fontsize=12, loc="upper left")
plt.grid(alpha=0.3)
plt.xlim(-2, 2)
plt.ylim(0, 8)
plt.xticks(ticks=np.arange(-2, 2.5, 0.5))

# Show the plot
plt.show()
```
