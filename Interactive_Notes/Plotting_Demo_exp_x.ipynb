```python
import numpy as np
from math import factorial
import matplotlib.pyplot as plt
```

```python
# Define Taylor Series approximations around x0 = 0
def taylor_exp_approximation_near_0(x, degree):
    x0 = 0  # Center point of the Taylor Series
    num_terms = degree + 1  # Number of terms to use in the approximation
    approximation = 0
    for n in range(num_terms):
        coefficient = 1 / factorial(n)
        term = coefficient * (x - x0)**n
        approximation += term
    return approximation
```

```python


# Define the exponential function
def exp_function(x):
    return np.exp(x)

# Define Taylor Series approximations around x0 = 0
def taylor_exp_approximation_near_0(x, degree):
    x0 = 0  # Center point of the Taylor Series
    num_terms = degree + 1  # Number of terms to use in the approximation
    approximation = 0
    for n in range(num_terms):
        coefficient = 1 / factorial(n)
        term = coefficient * (x - x0)**n
        approximation += term
    return approximation

# Generate x values
x = np.linspace(-2, 2, 500)

# Compute the true exponential function values
true_exp = exp_function(x)

# Plot the true exponential function
plt.figure(figsize=(10, 6))
plt.plot(x, true_exp, label="True $e^x$", color="black", linewidth=2)

# Plot Taylor Series approximations for various orders
degrees = [0, 1, 2, 3, 4]  # Orders of approximation
colors = ["red", "blue", "green", "orange", "purple"]

for degree, color in zip(degrees, colors):
    approximation = taylor_exp_approximation_near_0(x, degree)
    plt.plot(x, approximation, label=f"Degree {degree} Approximation", linestyle="--", color=color)

# Customize the plot
plt.title("Taylor Series Approximations of $e^x$ Near $x = 0$", fontsize=16)
plt.xlabel("$x$", fontsize=14)
plt.ylabel("$y$", fontsize=14)
plt.legend(fontsize=12, loc="upper left")
plt.grid(alpha=0.3)
plt.xlim(-2, 2)
plt.ylim(0, 8)
plt.xticks(ticks=np.arange(-2, 2.5, 0.5))

# Show the plot
plt.show()
```
